Clean the data

```python
file_path = "/Users/gemavidal/Box Sync/Vet/Projects and Papers/2020 YSM USAID Final Report/df_clinicalhistory_all.csv"
df = pd.read_csv(file_path, index_col = False)
```

```python
for i in df:
    print (df[i].value_counts())
```

```python
columns = ['medicationsincebirth','medicationsincebirth_dewormer','medicationsincebirth_oralelectrolytes','medicationsincebirth_vitamins',
           'signsofillness_diarrhea_notbloody', 'signsofillness_nasaldischarge']

for i in columns:
    df[i].replace({'No':'no',
                   'Ivermectin':'yes',
                   'multi-vitamin injection':'yes',
                   ' no':'no',
                   'unilateral':'yes',
                   'small amount, unilateral':'yes'}, inplace = True)
```

```python
# And create a column for the production system

def prod_system (row):
    if (row['district'] == 'Gondar'):
        return 'Peri-urban'
    if (row['district'] == 'Sululta'):
        return 'Peri-urban'
    if (row['district'] == 'Dalocha'):
        return 'Mixed-crop'
    if (row['district'] == 'Awash'):
        return 'Pastoral'
    if (row['district'] == 'Amibara'):
        return 'Pastoral'
```

```python
df['prod_system'] = df.apply(lambda row: prod_system(row), axis = 1)
```

```python
# Change all the ages into weeks as units
def label_ageweeks (row):
    if (row['age_units'] == 'weeks'):
        return row['age']
    if (row['age_units'] == 'months'):
        return row['age'] * 4
    if (row['age_units'] == 'days' and row['age'] >= 5.0 and row['age'] <=9):
        return 1
    if (row['age_units'] == 'days' and row['age'] < 5.0):
        return 0.5
    if (row['age_units'] == 'days' and row['age'] > 9.0):
        return row['age'] / 7 #because all the ages given the condition fit in there
```

```python
df['age_weeks'] = df.apply(lambda row: label_ageweeks(row), axis = 1)
```
