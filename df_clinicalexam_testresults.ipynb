```python
# And create a column for the production system

def prod_system (row):
    if (row['district'] == 'Gondar'):
        return 'Peri-urban'
    if (row['district'] == 'Sululta'):
        return 'Peri-urban'
    if (row['district'] == 'Dalocha'):
        return 'Mixed-crop'
    if (row['district'] == 'Awash'):
        return 'Pastoral'
    if (row['district'] == 'Amibara'):
        return 'Pastoral'
```

```python
df['prod_system'] = df.apply(lambda row: prod_system(row), axis = 1)
```

```python
df.loc[(df['age_units'] == 'days')]
```

```python
# Change all the ages into weeks as units
def label_ageweeks (row):
    if (row['age_units'] == 'weeks'):
        return row['age']
    if (row['age_units'] == 'months'):
        return row['age'] * 4
    if (row['age_units'] == 'days' and row['age'] >= 5.0 and row['age'] <=9):
        return 1
    if (row['age_units'] == 'days' and row['age'] < 5.0):
        return 0.5
    if (row['age_units'] == 'days' and row['age'] > 9.0):
        return row['age'] / 7 #because all the ages given the condition fit in there
```

```python
df['age_weeks'] = df.apply(lambda row: label_ageweeks(row), axis = 1)
```

```python
def dz_syndrome (row):
    if (row['pathogens'] == 'anaplasma_centrale_result' or 
        row['pathogens'] == 'anaplasma_marginale_result' or
        row['pathogens'] == 'babesia_bigemina_result' or
        row['pathogens'] == 'babesia_bovis_result'):
        return 'hemoparasites'
    if (row['pathogens'] == 'coccidia_result' or
        row['pathogens'] == 'result_ascaris' or
        row['pathogens'] == 'result_fasciola' or
        row['pathogens'] == 'result_monezia' or
        row['pathogens'] == 'result_paraphistom' or
        row['pathogens'] == 'result_schistosoma' or
        row['pathogens'] == 'result_strongly' or
        row['pathogens'] == 'result_trichuris'):
        return 'fecal_parasites'
    if (row['pathogens'] == 'corona_result' or
        row['pathogens'] == 'crypto_result' or
        row['pathogens'] == 'ecoli_result' or
        row['pathogens'] == 'rota_result'):
        return 'pathasure'
    if (row['pathogens'] == 'result_salmonella_enterica' or
        row['pathogens'] == 'BVDV_result'):
        return 'GI_pathogens'
    if (row['pathogens'] == 'm_hemolytica_result' or
        row['pathogens'] == 'p_multocida_result' or
        row['pathogens'] == 'result_ADV' or
        row['pathogens'] == 'result_BRSV' or
        row['pathogens'] == 'result_IBR' or
        row['pathogens'] == 'result_PIV3'):
        return 'respiratory_pathogens'
```

```python
df_pathogens['disease_syndrome'] = df_pathogens.apply(lambda row: dz_syndrome(row), axis = 1)
```
