```python
# Cleaning the data after append

columns = ['colostrumwithin24hours', 'continuousaccesstodamfirst24hours', 'dam_ageparturition_years',
           'dam_bcs1to5', 'parity', 'reasonnocolostrumfedwithin24hours', 'timeofbirth']

for i in columns:
    df_colostrum[i].replace({'unknown':np.nan,
                             'unkown':np.nan,
                             'NR':np.nan,
                             'Not applicable':np.nan,
                             'do not know':np.nan}, inplace = True)
```

```python
columns = ['colostrumfromstepmom','colostrumwithin24hours', 'continuousaccesstodamfirst24hours', 
           'sex', 'timeofbirth']

for i in columns:
    df_colostrum[i].replace({'Yes':'yes',
                             'No':'no',
                             'Female':'female',
                             'Male':'male',
                             'Male ':'male',
                             'male ':'male',
                             'Day':'day',
                             'Night':'night',
                             'Night ':'night'}, inplace = True)
```

```python
# Change the continous variables to floats

columns = ['dam_ageparturition_years', 'dam_bcs1to5', 'parity']

for i in columns:
    df_colostrum[i] = df_colostrum[i].astype(float)
```

```python
# fix 'parity'
columns = ['parity']
for i in columns:
    df_colostrum[i].replace({0.0:1}, inplace = True)
```

```python
# Cleaning dataframe

for i in df_colostrum:
    print (df_colostrum[i].value_counts())
```

```python
# fix 'dam_ageparturition_years'
columns = ['dam_ageparturition_years']
for i in columns:
    df_colostrum[i].replace({3.5 : 3,
                             5.5 : 5,
                             2.5 : 2,
                             2.7 : 2,
                             9.5 : 9,
                             4.8 : 4,
                             2.8 : 2,
                             2.9 : 2}, inplace = True)
```

```python
# Create a column for the production system
def prod_system (row):
    if (row['district'] == 'Gondar'):
        return 'Peri-urban'
    if (row['district'] == 'Sululta'):
        return 'Peri-urban'
    if (row['district'] == 'Dalocha'):
        return 'Mixed-crop'
    if (row['district'] == 'Awash'):
        return 'Pastoral'
    if (row['district'] == 'Amibara'):
        return 'Pastoral'
```

```python
df_colostrum['prod_system'] = df_colostrum.apply(lambda row: prod_system(row), axis = 1)
```

```python
# Change all the ages into weeks as units
def label_ageweeks (row):
    if (row['age_units'] == 'weeks'):
        return row['age']
    if (row['age_units'] == 'months'):
        return row['age'] * 4
    if (row['age_units'] == 'days' and row['age'] >= 5.0 and row['age'] <=9):
        return 1
    if (row['age_units'] == 'days' and row['age'] < 5.0):
        return 0.5
    if (row['age_units'] == 'days' and row['age'] > 9.0):
        return row['age'] / 7 #because all the ages given the condition fit in there
```

```python
df_colostrum['age_weeks'] = df_colostrum.apply(lambda row: label_ageweeks(row), axis = 1)
```
