## Cleaning data after the append

```python
for i in df_farm:
    print (df_farm[i].value_counts())
```

```python
columns = ['feedingfrequency', 'feedmilkormilkreplacer', 'naveldesinfection', 'supplementalfeed_frequency', 
           'typeofvaccinesreceived', 'vaccinationssincebirth', 'weaned_grazinghours_dryseason','weaned_grazinghours_rainyseason']

for i in columns:
    df_farm[i].replace({'unknown':np.nan,
                        'unkown':np.nan,
                        'unknown ':np.nan,
                        'NR':np.nan,
                        'sometimes':np.nan}, inplace = True)
```

```python
columns = ['breed','feedingfrequency','feedmilkormilkreplacer','sex']

for i in columns:
    df_farm[i].replace({'Cross':'Improved/Crossbred',
                        'local':'Local',
                        'cross':'Improved/Crossbred',
                        'Improved/Cross':'Improved/Crossbred',
                        'exotic':'Improved/Crossbred',
                        'twice a day':'Twice/day',
                        'twice a Day':'Twice/day',
                        'more than twice aday':'more than twice a day',
                        'once aday':'Once/day',
                        'once a Day':'Once/day',
                        'Thrice/day':'more than twice a day',
                        'Milk':'yes milk',
                        'female':'Female',
                        'female ':'Female',
                        'male':'Male',
                        'male ':'Male',
                        'Male ':'Male'}, inplace = True)
```

```python
for i in df_farm:
    print (df_farm[i].value_counts())
```

```python
columns = ['sire','supplementalfeed_frequency','watergiven_frequency']

for i in columns:
    df_farm[i].replace({'live cover':'Live Cover',
                        'Artificial Insemination':'AI',
                        'live Cover':'Live Cover',
                        'Live cover':'Live Cover',
                        'unknown':np.nan,
                        'twice a Day':'twice a day',
                        'more than twice a Day':'more than twice a day',
                        'once a Day':'once a day',
                        'do not know':np.nan}, inplace = True)
```

```python
columns = ['feedmilkormilkreplacer_liters']

for i in columns:
    df_farm[i].replace({'unknown':np.nan,
                        'unkown':np.nan,
                        'unknown ':np.nan,
                        'NR':np.nan}, inplace = True)
```

```python
columns = ['weaned_grazinghours_dryseason','weaned_grazinghours_rainyseason']

for i in columns:
    df_farm[i].replace({'10 up to 12 hrs':10.0,
                        '4 hours':4.0,
                        '1 hour':1.0,
                        '8 up to 10hrs ':8.0,
                        '8 up to 10hrs':8.0}, inplace = True)
```

```python
for i in df_farm:
    print (df_farm[i].value_counts())
```

```python
# Change the continous variables to floats

columns = ['weaned_grazinghours_dryseason', 'weaned_grazinghours_rainyseason']

for i in columns:
    df_farm[i] = df_farm[i].astype(float)
```

```python
# fix 'feedmilkormilkreplacer_liters'
columns = ['feedmilkormilkreplacer_liters']
for i in columns:
    df_farm[i].replace({'0.5' : 'half to 1 liter',
                        '0.3' : 'less than half liter',
                        '0.25' : 'less than half liter',
                        '0.2' : 'less than half liter',
                        '0.75' : 'half to 1 liter',
                        '0.4' : 'less than half liter',
                        '1' : 'half to 1 liter',
                        '0.1' : 'less than half liter',
                        '0.7' : 'half to 1 liter',
                        '5' : 'greater than 1 liter',
                        '0.55' : 'half to 1 liter'}, inplace = True)
```

```python
# Create a column for the production system
def prod_system (row):
    if (row['district'] == 'Gondar'):
        return 'Peri-urban'
    if (row['district'] == 'Sululta'):
        return 'Peri-urban'
    if (row['district'] == 'Dalocha'):
        return 'Mixed-crop'
    if (row['district'] == 'Awash'):
        return 'Pastoral'
    if (row['district'] == 'Amibara'):
        return 'Pastoral'
```

```python
df_farm['prod_system'] = df_farm.apply(lambda row: prod_system(row), axis = 1)
```
